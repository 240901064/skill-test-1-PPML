```python
#analyse student grades using numpy
import numpy as np

# Marks of students in Math, Science, English
marks = np.random.randint(50,100,(30,3))

subjects = ["Math","Science","English"]

# Average
avg = marks.mean(axis=0)

# Highest
high = marks.max(axis=0)

# Lowest
low = marks.min(axis=0)

for i in range(3):
    print(subjects[i])
    print("Average:",avg[i])
    print("Highest:",high[i])
    print("Lowest:",low[i])
    print()
```
