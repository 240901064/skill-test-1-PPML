```python
#total impedance of electrical circuit

import numpy as np
n = int(input("Enter number of components: "))
Z = []
for i in range(n):
    value = complex(input(f"Enter impedance Z{i+1} (example 10+5j): "))
    Z.append(value)
Z = np.array(Z)
choice = input("Enter connection type (series or parallel): ")
if choice == "series":
    Z_total = np.sum(Z)
    print("Total Series Impedance =", Z_total)
elif choice == "parallel":
    Z_total = 1 / np.sum(1/Z)
    print("Total Parallel Impedance =", Z_total)
else:
    print("Invalid connection type")
```
